## Exercise 3: Make nice figures

Now use `matplotlib` to reproduce as closely as you can figures 5 and 6 from the paper. This exercise is intended to get you to familiarize yourself with making nicely formatted `matplotlib` figures with multiple plots. Note that the plots in the paper are actually wrong!

```python
import matplotlib.pyplot as plt
import numpy as np

def create_subplot(ax, variable, xlabel, color_sig='red', color_bkg='blue'):
    ax.hist(df_sig[variable], bins=100, histtype='step',
            color=color_sig, label='Signal', density=True, stacked=True)
    ax.hist(df_bkg[variable], bins=100, histtype='step',
            color=color_bkg, label='Background', density=True, stacked=True)
    ax.set_xlabel(xlabel, fontsize=12)
    ax.set_ylabel("Arbitrary Units", fontsize=12)


fig5, axes5 = plt.subplots(2, 3, figsize=(15, 8))
fig5.suptitle('Low-level Features', fontsize=16)

# SubPlot: Low level Features
create_subplot(axes5[0, 0], 'l_1_pT', '$p_T(\ell_1)$ (GeV)')
create_subplot(axes5[0, 1], 'l_1_eta', '$\eta(\ell_1)$')
create_subplot(axes5[0, 2], 'l_1_phi', '$\phi(\ell_1)$')
create_subplot(axes5[1, 0], 'l_2_pT', '$p_T(\ell_2)$ (GeV)')
create_subplot(axes5[1, 1], 'l_2_eta', '$\eta(\ell_2)$')
create_subplot(axes5[1, 2], 'l_2_phi', '$\phi(\ell_2)$')

fig5.tight_layout(rect=[0, 0.03, 1, 0.95])
handles, labels = axes5[0, 0].get_legend_handles_labels()
fig5.legend(handles, labels, loc='upper right', fontsize=12)
plt.show()


fig6, axes6 = plt.subplots(2, 3, figsize=(15, 8))
fig6.suptitle('High-level Features', fontsize=16)

# SubPlots: High Level Features
create_subplot(axes6[0, 0], 'MET_rel', '$E^{rel}_T$ miss (GeV)')
create_subplot(axes6[0, 1], 'axial_MET', 'axial $E_T$ miss (GeV)')
create_subplot(axes6[0, 2], 'M_TR_2', '$M_{T2}$ (GeV)')
create_subplot(axes6[1, 0], 'M_R', '$M_R$ (GeV)')
create_subplot(axes6[1, 1], 'S_R', '$S_R$')
create_subplot(axes6[1, 2], 'R', '$R$')

fig6.tight_layout(rect=[0, 0.03, 1, 0.95])
handles, labels = axes6[0, 0].get_legend_handles_labels()
fig6.legend(handles, labels, loc='upper right', fontsize=12)
plt.show()
```

```python
## 4.1 a solution
import matplotlib.pyplot as plt
def create_pair_plots(df, variables):
    num_vars = len(variables)
    fig, axes = plt.subplots(num_vars, num_vars, figsize=(12, 12))

    data = [df[var].values for var in variables]  

    for i in range(num_vars):
        for j in range(num_vars):
            if i == j:
                # Diagonal: Histogram
                axes[i, j].hist(data[i], bins=30) 
                axes[i, j].set_title(variables[i])
            else:
                # Off-diagonal: Scatter plot
                axes[i, j].scatter(data[j], data[i], alpha=0.5)  

            # axis labels
            if i == num_vars - 1:
                axes[i, j].set_xlabel(variables[j])
            if j == 0:
                axes[i, j].set_ylabel(variables[i])
                
    plt.suptitle("Pair Plot of Selected Features", fontsize=16)
    plt.tight_layout()
    plt.show()
```

```python
### part a testing

# low-level features:
print("Pair plot for Low-Level Features:")
create_pair_plots(df, RawNames)

# high-level features:
print("Pair plot for High-Level Features:")
create_pair_plots(df, FeatureNames)
```

```python
## 4.1 b solution

def speed_pair_plots(df, variables):
    num_vars = len(variables)
    fig, axes = plt.subplots(num_vars, num_vars, figsize=(12, 12))

    data = [df[var].values for var in variables]  

    for i in range(num_vars):
        for j in range(num_vars):
            if i == j:
                axes[i, j].hist(data[i], bins=30) 
                axes[i, j].set_title(variables[i])
            else:
                axes[i, j].scatter(data[j], data[i], alpha=0.5)  
            if i == num_vars - 1:
                axes[i, j].set_xlabel(variables[j])
            if j == 0:
                axes[i, j].set_ylabel(variables[i])

    plt.suptitle("Optimized Pair Plot of Selected Features", fontsize=16)
    plt.tight_layout()
    plt.show()
```

```python
### 4.1 b testing

# low-level features:
print("Pair plot for Low-Level Features:")
speed_pair_plots(df, RawNames)

# high-level features:
print("Pair plot for High-Level Features:")
speed_pair_plots(df, FeatureNames)
```

```python
# part b
import numpy as np
# All observables
all_observables = VarNames[1:]  # Exclude the 'signal' column
cov_matrix_all = np.cov(df[all_observables], rowvar=False)
corr_matrix_all = np.corrcoef(df[all_observables], rowvar=False)

# Low-level features
cov_matrix_low = np.cov(df[RawNames], rowvar=False)
corr_matrix_low = np.corrcoef(df[RawNames], rowvar=False)

# High-level features
cov_matrix_high = np.cov(df[FeatureNames], rowvar=False)
corr_matrix_high = np.corrcoef(df[FeatureNames], rowvar=False)
```

```python
# Part c
import tabulate
from IPython.display import HTML, display

def format_matrix(matrix, headers):
    """Formats a matrix for tabulate with headers and significant figures."""
    formatted_matrix = []
    for i, row in enumerate(matrix):
        formatted_row = [headers[i]] + [f"{val:.3f}" for val in row]  # 3 significant figures
        formatted_matrix.append(formatted_row)
    return formatted_matrix

table_cov_all = format_matrix(cov_matrix_all, all_observables)
table_corr_all = format_matrix(corr_matrix_all, all_observables)
table_cov_low = format_matrix(cov_matrix_low, RawNames)
table_corr_low = format_matrix(corr_matrix_low, RawNames)
table_cov_high = format_matrix(cov_matrix_high, FeatureNames)
table_corr_high = format_matrix(corr_matrix_high, FeatureNames)


### Part c test
print("Covariance: All Features")
display(HTML(tabulate.tabulate(table_cov_all, tablefmt='html', headers=[''] + all_observables)))
print("Correlation: All Features")
display(HTML(tabulate.tabulate(table_corr_all, tablefmt='html', headers=[''] + all_observables)))
print("Covariance: Low-Level Features")
display(HTML(tabulate.tabulate(table_cov_low, tablefmt='html', headers=[''] + RawNames)))
print("Correlation: Low-Level Features")
display(HTML(tabulate.tabulate(table_corr_low, tablefmt='html', headers=[''] + RawNames)))
print("Covariance: High-Level Features")
display(HTML(tabulate.tabulate(table_cov_high, tablefmt='html', headers=[''] + FeatureNames)))
print("Correlation: High-Level Features")
display(HTML(tabulate.tabulate(table_corr_high, tablefmt='html', headers=[''] + FeatureNames)))
```

```python
# Part d
def compute_and_display_matrices(df, variables, title):
    """Computes and displays covariance and correlation matrices using tabulate."""
    cov_matrix = np.cov(df[variables], rowvar=False)
    corr_matrix = np.corrcoef(df[variables], rowvar=False)

    table_cov = format_matrix(cov_matrix, variables)
    table_corr = format_matrix(corr_matrix, variables)

    display(HTML(f"<h3>{title}</h3>"))  # Add a title
    display(HTML(tabulate.tabulate(table_cov, tablefmt='html', headers=[''] + variables)))
    display(HTML(tabulate.tabulate(table_corr, tablefmt='html', headers=[''] + variables)))

### part d test
compute_and_display_matrices(df, all_observables, "All Observables")
compute_and_display_matrices(df, RawNames, "Low-level Features")
compute_and_display_matrices(df, FeatureNames, "High-level Features")
```
